# Sampling methods applied
So far there have been two samples created from separate sampling methods: 
* **Simple random sampling**: Median housing prices were selected randomly from the entire dataset. The probability of any median house price being selected in the sample was equal. This was used to create the sample set of *train_data* and *test_data*
* **Stratified random sampling**: Median housing prices were first stratified (categorised) into various discrete categories, and then randomly selected. The number of median house prices selected from that category reflected the size of that category. This method created the sample sets *stratified_train* and *stratified_test*

### Comparison of sampling methods

```python
original_median_value_proportions = data['income_cat'].value_counts() / len(data)

# creating training and testing samples through simple sampling
simple_train, simple_test = train_test_split(data, test_size = 0.2, random_state = 42)

simple_median_value_proportions = simple_train['income_cat'].value_counts() / len(simple_train)
stratified_median_value_proportions = stratified_train['income_cat'].value_counts() / len(stratified_train)

comparison = pd.DataFrame({
    "original_proportions": original_median_value_proportions,
    "simple_random_sampling": simple_median_value_proportions,
    "stratified_random_sampling": stratified_median_value_proportions
})
comparison['simple_sampling_error'] = (comparison['original_proportions'] - comparison['simple_random_sampling']) * 100 / comparison['original_proportions']
comparison['stratified_sampling_error'] = (comparison['original_proportions'] - comparison['stratified_random_sampling']) * 100 / comparison['original_proportions']
comparison[["simple_sampling_error","stratified_sampling_error"]]
```

The results above show that the stratified sampling method produces a smaller error than simple random sampling.

# Preparing the data for a machine learning algorithm

Creating a clean training dataset, separating the label from the predictors:

```python
housing = stratified_train.drop('median_house_value', axis=1)
housing_label = stratified_train['median_house_value'].copy()
```

```python
#finding the na values in the housing dataframe:
housing_na = housing[housing.isnull().any(axis=1)]
housing_na.head()
```

```python
from sklearn.impute import SimpleImputer
#Imputer is used to fill all na values in a dataset.

imputer = SimpleImputer(strategy='median')
```

```python
#Splitting training set into numerical features and categorical features to impute missing values:
housing_numerical = housing.drop('ocean_proximity', axis=1)
housing_categorical = housing[['ocean_proximity']]
```

### Preprocessing the numerical features:
* Replacing null values with median

```python
imputer.fit(housing_numerical)
imputer.statistics_

#returns a numpy array:
X = imputer.transform(housing_numerical)
#create a dataframe from the generated array with null values replaced:
housing_num_imputed = pd.DataFrame(X, columns=housing_numerical.columns, index=list(housing_numerical.index.values))
#the same needs to be done with the categorical features:
```

### Preprocessing the categorical features:
* Replacing categories with numerical values

```python
from sklearn.preprocessing import OneHotEncoder

categorical_encoder = OneHotEncoder()
housing_categorical_1hot= categorical_encoder.fit_transform(housing_categorical)
categorical_encoder.categories_
```

### Creating a class to add some transformed features

```python
from sklearn.base import BaseEstimator, TransformerMixin

rooms_ix, bedrooms_ix, population_ix, household_ix = 3,4,5,6

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room=True):
        self.add_bedrooms_per_room = add_bedrooms_per_room
    def fit(self, X, y=None):
        return self
    def transform(self, X, y=None):
        rooms_per_household = X[:, rooms_ix] / X[:, household_ix]
        population_per_household = X[:, population_ix] / X[:, household_ix]
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            return np.c_[X, rooms_per_household, population_per_household, bedrooms_per_room]
        else: 
            return np.c_[X, rooms_per_household, population_per_household]
        
attribute_adder = CombinedAttributesAdder(add_bedrooms_per_room=False)

housing_new_attributes = attribute_adder.transform(housing.values)
#the field above is a series of arrays, this can be converted back into a dataframe

housing_new_attributes = pd.DataFrame(
    housing_new_attributes,
    columns = list(housing.columns) + ['rooms_per_household','population_per_household'])

#housing_new_attributes is now a dataframe with new attributes added:
housing_new_attributes.head()
```

### Creating a pipeline for calculating numerical attributes

Numerical data needs to be standardized. Machine learning algorithms do not perform well with large variations in scales. Standardizing reduces all numerical features to a value between 0 and 1. 

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('attribute_adder', CombinedAttributesAdder()),
    ('std_scaler', StandardScaler()),
])

housing_numerical_tr = num_pipeline.fit_transform(housing_numerical)
housing_numerical_tr
```

The Pipeline above applies transformations to only numerical features. The ColumnTransformer can act as a pipeline for both numerical and categorical features:

```python
from sklearn.compose import ColumnTransformer

num_attributes = list(housing_num_imputed)
cat_attributes = ['ocean_proximity']

total_pipeline = ColumnTransformer([
    ("num", num_pipeline, num_attributes),
    ("cat", OneHotEncoder(), cat_attributes)
])

housing_final = total_pipeline.fit_transform(housing)
housing_final[0]
```

# Linear Regression model

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(housing_final, housing_label)
```

The model has been created and trained with the housing data. Now it can be used for prediction:

```python
sample_features = housing.iloc[:5]
sample_label = housing_label.iloc[:5]
transformed_sample = total_pipeline.transform(sample_features)
print('Predicted median house values: ', lin_reg.predict(transformed_sample))
```

The values above are the predicted median house values for the first 5 rows. Lets compare them to the actual house values:

```python
print('Actual median house values: ', list(sample_label))
```

The predicted values are not quite as accurate as expected. The Root Mean Squared Error (RMSE) can be determined for the full training dataset: 

```python
from sklearn.metrics import mean_squared_error

#running the pipeline for the full dataset:
lin_predicted_values = lin_reg.predict(housing_final)
lin_mse = mean_squared_error(housing_label, lin_predicted_values)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
housing_label.describe()
```

The RMSE is approximately $68,628. A better model will be needed: 

# Decision Tree Regressor model

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor(random_state=42)
tree_reg.fit(housing_final, housing_label)
```

```python
tree_predicted_values = tree_reg.predict(transformed_sample)
tree_mse = mean_squared_error(tree_predicted_values, sample_label)
tree_rmse = np.sqrt(tree_mse)
tree_rmse
```

The RMSE is 0 as a result of the Decision Tree Regressor. It's quite likely that the model has overfit the data. A better way to test this is to use **K-fold cross-validation**: 

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(tree_reg, housing_final, housing_label, scoring='neg_mean_squared_error', cv=10)
final_tree_score = np.sqrt(-scores)
final_tree_score
```

We can see that the **decision tree regressor** looks slightly worse than the **linear regression** model. To confirm this, the same *K-fold cross-validation* should be applied to the linear regression model:

```python
lin_reg_score = cross_val_score(lin_reg, housing_final, housing_label, scoring='neg_mean_squared_error', cv=10)
lin_reg_final_score = np.sqrt(-lin_reg_score)
lin_reg_final_score
```

```python
model_performance = pd.DataFrame(
    [["linear_regression", lin_reg_final_score.mean(), lin_reg_final_score.std()],
    ["decision_tree_regressor", final_tree_score.mean(), final_tree_score.std()]],
    columns = ['model','rmse_mean','rmse_std']
)
model_performance.sort_values(by='rmse_mean',ascending=True)
```

The DataFrame above shows that the **linear regression** model performs better that the **decision tree regressor**. Another model that could be used is the **random forest regressor**.

# Random Forest Regressor model

```python
from sklearn.ensemble import RandomForestRegressor

forest_reg = RandomForestRegressor(n_estimators=10, random_state=42)
forest_reg.fit(housing_final, housing_label)
```

```python
forest_predicted_values = forest_reg.predict(transformed_sample)
forest_mse = mean_squared_error(forest_predicted_values, sample_label)
forest_rmse = np.sqrt(forest_mse)
forest_rmse
```

Again, this would be more reliable if *K-fold Cross validation* was used:

```python
forest_cross_val = cross_val_score(forest_reg, housing_final, housing_label, scoring='neg_mean_squared_error', cv=10)
forest_cross_val_score = np.sqrt(-forest_cross_val)
forest_cross_val_score
```

```python
model_performance.append(pd.DataFrame(
    [['random_forest_regressor',forest_cross_val_score.mean(),forest_cross_val_score.std()]],
    columns=['model','rmse_mean','rmse_std'],
    index = [2]
)).sort_values(by='rmse_mean',ascending=True)
```

So far, the **Random Forest Regressor** has produced outputs with the lowest root mean squared error.
